```python
from sklearn.metrics import mean_squared_error
```

```python
def get_slope(day, koutei):
    X = day.reshape(len(day), 1)
    y = koutei

    lr.fit(X, y)
    slope = lr.coef_[0]
    inter = lr.intercept_
    mse = mean_squared_error(koutei, inter + slope*day)

    return slope, inter, mse
```

```python
def get_eig(X, y):
#     X = [[float(x)] for x in day]
#     y = koutei

    data = np.array([X, y]).T
    pca.fit(data)
    cmp = pca.components_
    cov = pca.get_covariance()
#     eig = LA.eig(cov)[1][1, 0] / LA.eig(cov)[1][0, 0]
    
#     return eig, cov, cmp

    slope = cmp[0, 1] / cmp[0, 0]
    var = cov[1, 1] / cov[0, 0]
    return slope, var
```
